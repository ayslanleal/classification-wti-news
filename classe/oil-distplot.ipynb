```python
import pandas as pd
import pandas_datareader.data as web
import yfinance as yf
import numpy as np
import seaborn as sns
import math
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
yf.pdr_override()
```

```python
oil = web.get_data_yahoo('CL=F', start = '2019-01-01', end = '2020-02-28')
```

```python
array = np.array(oil['Close'])
array2 = np.array(oil['Close'][1:])
evol = []
for i in range(len(array2)):
    aux = ((array2[i] - array[i])/array2[i]) * 100
    evol.append(aux)
evol.insert(0,0)

oil['% Evol'] = evol

oil_evol = pd.DataFrame(oil['% Evol'])
```

```python
oil['Close'].plot(figsize = (10,5))
plt.title('% de Evolução diário WTI jan-2019 até fev-2020')
plt.show()
```

![Distribuição normal](https://www.monolitonimbus.com.br/wp-content/uploads/2016/10/Standard_deviation_diagram.png)

```python
std_soma = oil_evol.mean()+oil_evol.std()
std_sub = oil_evol.mean()-oil_evol.std()
sns.distplot(oil_evol)
plt.title('Distribuição % de evolução diário WTI jan-2019/fev-2020')
plt.show()
```

```python
print('Media {}'.format(oil_evol.mean()))
print('Media+DesvioPadrao {}'.format(std_soma))
print('Media-DesvioPadrao {}'.format(std_sub))
```
